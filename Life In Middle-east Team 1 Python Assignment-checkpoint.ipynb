```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm


"""An amazing challenge was presented to us which was to analyze the middleast Pennisular region and the countries that
make it up. We were given a dataset from the world bank that represented the state of various metrics in terms of
health,education,military spending, GDP, Life expectancy etc for the year 2017 showcasing the countries in the middleeast
We started by adding the missing data that we found from various sources such as world bank, united nations etc. We analysed
the data by making correlation matrix, boxplots, multi variable regression, histograms and regression plots.
We were able to come up with great insights from the data that we were able to back up by news articles
that explained the outliers and showed a snapshot about the standard of livng in the middle-east peninsula."""

file = "Working dataset - Python.xlsx"
Middleast = pd.read_excel(file)#reading the file

#Copying the data into dataframe for manipulation
Middleast_data = pd.DataFrame.copy(Middleast)

#Describes the Data in terms statistical terms such as mean standard deviation count etc.
Middleast_data.describe()

#narrowing down to exact dataframe we would need
Ar_of_interest = Middleast_data['Cool Name'] == 'Wheeljack' 
wheeljack = Middleast_data[Ar_of_interest]

#looking at the data that we selected
print(wheeljack.iloc[: , 8:64].isnull().sum())
print(type(wheeljack))

#barplot explaining adolescent fertility rate
x = wheeljack['Country Code']
y = wheeljack['Adolescent fertility rate (births per 1,000 women ages 15-19)']

plt.xlabel('Country Code')
plt.ylabel('Birth per 1000 women (ages 15-19)')
plt.title('Adolescent Fertility Rate')
plt.tight_layout()
plt.bar(x, y, color = 'red')
plt.savefig('Ad_Fertility.png')
plt.show()

#Urban vs Rural Bar plot

x_a = wheeljack['Country Code']
y = wheeljack['GDP (current US$)']

plt.xlabel('Country Code')
plt.ylabel('GDP (current US$)')
plt.title('GDP of Countries')
plt.tight_layout()
plt.bar(x, y, color = 'green')
plt.savefig('Urban-vs-Rural.png')
plt.show()

#Access to electricity barplot rural vs urban
n_groups = 15

x_b = wheeljack['Access to electricity, rural (% of rural population)']
x_c = wheeljack['Access to electricity, urban (% of urban population)']

fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.3

graph1 = plt.bar(index, x_b, bar_width, alpha = opacity, color = 'blue',
                label = 'Rural Populations Access to Electricity')
graph2 = plt.bar(index+bar_width, x_c, bar_width, color = 'blue',
                label = 'Urban Populations Access to Electricity')

plt.xlabel('Country Code')
plt.ylabel('% of Population with access to electricity')
plt.xticks(index+bar_width, (wheeljack['Country Code']))
plt.legend(loc = 'lower right')
plt.tight_layout()
plt.savefig('Rural Vs Urban Populations Access to electricity.png')
plt.show()
#Age dependancy ratio bar plot
N = 15
young = wheeljack['Age dependency ratio, young (% of working-age population)']
old = wheeljack['Age dependency ratio, old (% of working-age population)']

ind = np.arange(N)
width = 0.50

graph3 = plt.bar((wheeljack['Country Code']), young, width, color = 'green')
graph4 = plt.bar((wheeljack['Country Code']), old, width, color = 'brown', bottom = young.tolist())

plt.xlabel('Country Code')
plt.ylabel('% of Working-Age Population')
plt.title('Dependecy on the Total Working-Age Population')
plt.xticks(ind, (wheeljack['Country Code']))
plt.yticks(np.arange(0, 75, 10))
plt.legend((graph3[0], graph4[0]), ('Young', 'Old'))
plt.savefig('Dependecy on the Total Working Age Population.png')
plt.show()

#birth vs death rate bar plot
N = 15
Birth = wheeljack['Birth rate, crude (per 1,000 people)']
Death = wheeljack['Death rate, crude (per 1,000 people)']

ind = np.arange(N)
width = 0.50

graph5 = plt.bar((wheeljack['Country Code']), Birth, width, color = 'orange')
graph6 = plt.bar((wheeljack['Country Code']), Death, width, color = 'black', bottom = Birth.tolist())

plt.xlabel('Country Code')
plt.ylabel('Birth vs Death')
plt.title('Birth & Death rates')
plt.xticks(ind, (wheeljack['Country Code']))
plt.yticks(np.arange(0, 75, 10))
plt.legend((graph5[0], graph6[0]), ('Birth', 'Death'))
plt.savefig('Birth-Death.png')
plt.show()

#scatter plot and regression plot for death rate and birth rate
wheeljack.plot(kind='scatter', x='GDP (current US$)', y='Death rate, crude (per 1,000 people)', color='Blue'); 
plt.show()
plt.savefig('GDP-Deaht.png')
wheeljack.plot(kind='scatter', x='GDP (current US$)', y='Birth rate, crude (per 1,000 people)', color='Red'); 
plt.show()
plt.savefig('GDP-Birth.png')
ax = sns.regplot(x='GDP (current US$)', y='Birth rate, crude (per 1,000 people)', data=wheeljack)
palette = sns.color_palette("Paired", 10)
sns.set_palette(palette)
plt.show()
plt.savefig('Regplot-GDP-Birth.png')

#Print GDP in numeric values
print(wheeljack[['GDP (current US$)','Country Name']])
plt.tight_layout()
plt.show()
plt.savefig('country-data.png')


#Government expenditure on education
sns.set_style('whitegrid')
show = wheeljack.groupby(["Country Name"])["Government expenditure on education, total (% of government expenditure)"].aggregate(np.median).reset_index().sort_values("Government expenditure on education, total (% of government expenditure)")
sns.barplot(x="Government expenditure on education, total (% of government expenditure)", y='Country Name', data = wheeljack, order=show['Country Name'], orient = 'h', color = 'blue', alpha=0.5)
plt.title('Government expenditure on education, total (% of government expenditure)')
plt.xlabel('')
plt.ylabel('')
plt.tight_layout()
plt.savefig('Barplot education expenditure.png')
plt.show()
#Urban and rural population
urban_population=wheeljack["Urban population (% of total population)"]
rural_population=wheeljack["Rural population (% of total population)"]
plt.bar(wheeljack['Country Name'], rural_population, color='red', alpha=0.3)
plt.bar(wheeljack['Country Name'], urban_population, color='red', alpha=0.6, bottom=rural_population.tolist())
plt.title("Percentage of total population - split in urban and rural")
plt.legend(['Rural', 'Urban'])
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig('Barplot, Percentage of total population - split in urban and rural.png')
plt.show()

#births attended by skilled staff
print(wheeljack["Births attended by skilled health staff (% of total)"])
plt.savefig('Births attended by skilled staff.png')

#male-female percentage split
gen=['% Male','% Female']
Per_mal=wheeljack['Population, male']/wheeljack['Population, total']*100
Per_fem=wheeljack['Population, female']/wheeljack['Population, total']*100
plt.bar(wheeljack['Country Code'],Per_mal)
plt.bar(wheeljack['Country Code'],Per_fem,bottom=Per_mal.tolist())
plt.title('Population split by gender')
plt.legend(gen, loc=1)
plt.tight_layout()
plt.savefig('Population split by gender')
plt.show()

#Multi variable plot for births attended by skilled staff, life expectancy at birth,Adolescent fertility rate
g = sns.pairplot(wheeljack,height=5,palette="husl",hue="Country Code",
                     x_vars=['Births attended by skilled health staff (% of total)', 'Life expectancy at birth, total (years)','Adolescent fertility rate (births per 1,000 women ages 15-19)'],
                     y_vars=['Births attended by skilled health staff (% of total)', 'Life expectancy at birth, total (years)','Adolescent fertility rate (births per 1,000 women ages 15-19)'])
plt.savefig('Multivariableplot.png')
plt.show()


#GDP Growth perentage change
x = wheeljack['Country Code']
y = wheeljack['Population growth (annual %)']


Country=['non-conflict']
plt.xlabel('Country Code')
plt.ylabel('Population growth (annual %)')
plt.title('Population growth')
plt.tight_layout()
plt.bar(x, y, color = ('blue','blue','red','red','red','blue','red','blue','blue','blue','red','blue','blue','red','red'), edgecolor = 'blue' )
plt.savefig('Population growth.png')
plt.legend(Country, loc=3)
plt.show()

# Conflicted vs non conflicted countries

x = wheeljack['Country Code']
y = wheeljack['GDP growth (annual %)']

plt.xlabel('Country Code')
plt.ylabel('GDP growth (annual %)')
plt.title('Oil producer - GDP Growth')
plt.tight_layout()
plt.bar(x, y, color = ('blue','blue','green','blue','blue','green','blue','green','blue','green','blue','blue','green','blue','blue'))
plt.savefig('GDP Growth Oil producer.png')
plt.show()

#boxplot gdp growth in annual percentage change
wheeljack.boxplot(column='GDP growth (annual %)')
plt.title("GDP % Change")
plt.xlabel('')
plt.ylabel('')
plt.tight_layout()
plt.savefig('Boxplot, Prevalence of undernourishment (% of population).png')
plt.show()

#histogram for employment in agriculture
plt.hist(x="Employment in agriculture (% of total employment) (modeled ILO estimate)",
data=wheeljack)
plt.xlabel('Employment in agriculture %')

#Boxplot Adolescent fertility rate
wheeljack.boxplot(column='Adolescent fertility rate (births per 1,000 women ages 15-19)')
plt.title("Adolescent fertility rate (births per 1,000 women ages 15-19)")
plt.xlabel('')
plt.ylabel('')
plt.tight_layout()
plt.savefig('Boxplot, Adolescent fertility rate (births per 1,000 women ages 15-19).png')
plt.show()

#Boxplot Access to electricity rural
wheeljack.boxplot(column='Access to electricity, rural (% of rural population)')
plt.title("Access to electricity, rural (% of rural population)")
plt.xlabel('')
plt.ylabel('')
plt.tight_layout()
plt.savefig('Boxplot, Access to electricity, rural (% of rural population).png')
plt.show()

#Boxplot births attended by skilled staff
wheeljack.boxplot(column='Births attended by skilled health staff (% of total)')
plt.title("Births attended by skilled health staff (% of total)")
plt.xlabel('')
plt.ylabel('')
plt.tight_layout()
plt.savefig('Boxplot, Births attended by skilled health staff (% of total).png')
plt.show()

#Boxplot Current GDP in USD
wheeljack.boxplot(column='GDP (current US$)')
plt.title("GDP (current US$)")
plt.xlabel('')
plt.ylabel('')
plt.tight_layout()
plt.savefig('Boxplot, GDP (current US$).png')
plt.show()

#Boxplot life expectancy at birth
wheeljack.boxplot(column='Life expectancy at birth, total (years)')
plt.title("Life expectancy at birth, total (years)")
plt.xlabel('')
plt.ylabel('')
plt.tight_layout()
plt.savefig('Life expectancy at birth, total (years)')
plt.show()

#Boxplot Military expenditure % of GDP
wheeljack.boxplot(column='Military expenditure (% of GDP)')
plt.title("Military expenditure (% of GDP)")
plt.xlabel('')
plt.ylabel('')
plt.tight_layout()
plt.savefig('Boxplot, Military expenditure (% of GDP).png')
plt.show()

#Boxplot population above 65 years old in %
wheeljack.boxplot(column='Population ages 65 and above (% of total population)')
plt.title("Population ages 65 and above (% of total population)")
plt.xlabel('')
plt.ylabel('')
plt.tight_layout()
plt.savefig('Boxplot, Population ages 65 and above (% of total population).png')
plt.show()


#Boxplot of population density
wheeljack.boxplot(column='Population density (people per sq. km of land area)')
plt.title("Population density (people per sq. km of land area)")
plt.xlabel('')
plt.ylabel('')
plt.tight_layout()
plt.savefig('Boxplot, Population density (people per sq. km of land area).png')
plt.show()


#Boxplot Population growth annual %
wheeljack.boxplot(column='Population growth (annual %)')
plt.title("Population growth (annual %)")
plt.xlabel('')
plt.ylabel('')
plt.tight_layout()
plt.savefig('Boxplot, Population growth (annual %).png')
plt.show()


#Boxplot Rural Population as percentage of total population
wheeljack.boxplot(column='Rural population (% of total population)')
plt.title("Rural population (% of total population)")
plt.xlabel('')
plt.ylabel('')
plt.tight_layout()
plt.savefig('Boxplot, Rural population (% of total population).png')
plt.show()


#Correlation Matrix
f = plt.figure(figsize=(19, 15))
plt.matshow(wheeljack.corr(), fignum=f.number)
plt.xticks(range(wheeljack.shape[1]), wheeljack.columns, fontsize=14, rotation=45)
plt.yticks(range(wheeljack.shape[1]), wheeljack.columns, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix', fontsize=16);
plt.savefig('Correlation Matrix.png')
plt.show()



# End of Code 
```
